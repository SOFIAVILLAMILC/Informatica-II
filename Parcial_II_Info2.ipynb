```python
import matplotlib.pyplot as plt
import scipy.io
class Electromiografía:
    def __init__(self,filename):
        data = scipy.io.loadmat(filename)
        self.signal = data['datos']
        self.canales = self.signal.shape[0]
        self.puntos = self.signal.shape[1]

    def mostrarInformacion(self):
        print(f"La señal tiene {self.canales} canales y {self.puntos} puntos de tiempo")

    def graficar(self, limites):
        filas = self.canales // 2
        fig, axes = plt.subplots(filas, 2, figsize=(10, 5))
        Iterador = axes.flat

        for i, ax in enumerate(Iterador):
            ax.plot(self.signal[i], color = "yellow")
            ax.set_title(f"Canal {i+1}")
            ax.set_xlabel("Tiempo")
            ax.set_ylabel("Amplitud")
            ax.grid(True)
            ax.set_xlim(limites)

        plt.tight_layout()
        plt.show()

    def grafica_promedio(self):
        señal_promedio = np.mean(self.signal, axis=0)
        
        plt.figure(figsize=(10, 5))
        plt.plot(señal_promedio, color='pink', linewidth=3)
        plt.title("Señal promedio")
        plt.xlabel("Tiempo")
        plt.ylabel("Amplitud")
        plt.grid(True)
        plt.show()



archivo = Electromiografía(r"C:\Users\sofia\OneDrive\Escritorio\Info 2024\Info-2\Info-2\Unidad 2\Parcial II\archivo1.mat")
archivo.mostrarInformacion()
archivo.graficar((0,100))
archivo.grafica_promedio()
archivo2 = Electromiografía(r"C:\Users\sofia\OneDrive\Escritorio\Info 2024\Info-2\Info-2\Unidad 2\Parcial II\archivo2.mat")
archivo2.mostrarInformacion()
archivo2.graficar((0,100))
archivo2.grafica_promedio()
```

2. (20%) Utilice la librería NumPy para generar una serie de señales sinusoidales con las siguientes características:

- **Frecuencia de la señal: 1 Hz.**
- **Número de muestras: 1000.**
- **Rango de amplitud: entre 1 y 5 unidades.**

Genere un total de 5 señales sinusoidales, cada una con una amplitud aleatoria dentro del rango especificado.

Utilice la librería Matplotlib para graficar las 5 señales sinusoidales en una misma figura y cada señal debe estar etiquetada con su respectiva amplitud en la leyenda.

Incluya etiquetas adecuadas en los ejes x e y de la figura, así como un título descriptivo.

```python
import numpy as np
import matplotlib.pyplot as plt
#señal sinusoidal: funcion seno
x = np.linspace(0,2*np.pi,1000) #inicio en 0, fin en 2pi que es la frecuencia y son 1000 muestras
amplitud = 1 + (4 * np.random.rand(5)) #random entre 0 y 1*4+1
y1 = amplitud[0]*np.sin(x)
y2 = amplitud[1]*np.sin(x) 
y3 = amplitud[2]*np.sin(x) 
y4 = amplitud[3]*np.sin(x)
y5 = amplitud[4]*np.sin(x)

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x,y1, label = f"Seña sinusoidal de amplitud {amplitud[0]}", color = "purple", linewidth = 3) 
ax.plot(x,y2, label = f"Seña sinusoidal de amplitud {amplitud[1]}", color = "pink", linewidth = 3) 
ax.plot(x,y3, label = f"Seña sinusoidal de amplitud {amplitud[2]}", color = "red", linewidth = 3) 
ax.plot(x,y4, label = f"Seña sinusoidal de amplitud {amplitud[3]}", color = "blue", linewidth = 3) 
ax.plot(x,y5, label = f"Seña sinusoidal de amplitud {amplitud[4]}", color = "green", linewidth = 3) 


ax.set_xlabel("Frecuencia 2π / 1Hz") 
ax.set_ylabel("Amplitud") 
ax.set_title("Señales sinusoidales") 
ax.legend() 
ax.grid()





plt.show() 
```
